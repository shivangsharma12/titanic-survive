```python
plt.figure(figsize=(5,5))
plt.bar(titanic_data['Survived'].value_counts().index, titanic_data['Survived'].value_counts(), color=["r","g"])
plt.xlabel('Survived')  # 0 = Did not survive, 1 = Survived
plt.ylabel('Count')
plt.title('Survival Counts in Titanic Dataset')
plt.xticks(ticks=[0, 1], labels=['Did Not Survive', 'Survived'])  
plt.show()
```
